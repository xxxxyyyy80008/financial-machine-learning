```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
from datetime import datetime, timedelta, time, date
```

```python
import yfinance as yf
```

```python
#read 544 ticker list
df_tickers = pd.read_csv('data/544tickers.csv')
```

```python
df_tickers.shape
```

```python
df_tickers.head()
```

```python
ticker_list = df_tickers['Symbol'].unique()
print(len(ticker_list))
```

```python
ticker_list[:3]
```

```python
dest_path = Path('data/raw_price')
print(dest_path, dest_path.exists())

if not dest_path.exists():
    dest_path_t = dest_path.joinpath('tmp')
    dest_path_t.parent.mkdir(parents=True, exist_ok=True)

print(dest_path, dest_path.exists())
```

```python
for ticker in ticker_list:
    cur_data = yf.Ticker(ticker)
    hist = cur_data.history(period="max")
    print(datetime.now(), ticker, hist.shape, hist.index.min(), hist.index.max())
    hist.to_csv(dest_path.joinpath(f'{ticker}.csv'), index=True, sep='|')
```
